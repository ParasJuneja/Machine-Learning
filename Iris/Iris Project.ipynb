```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import numpy as np
```

```python
df = pd.read_csv('iris.data',header = -1)
df.head()
```

```python
cols_name = ['sepal_len', 'sepal_width', 'petal_len', 'petal_width', 'species']
df.columns = cols_name
```

```python
df.hist(edgecolor = 'black', linewidth = 2, figsize = (12,8))
plt.show()
```

```python
sns.pairplot(df, hue = 'species', size = 4)
```

```python
plt.figure(figsize = (15, 10))
plt.subplot(2,2,1)
sns.violinplot(x = 'species', y = 'sepal_len', data = df)
plt.subplot(2,2,2)
sns.violinplot(x = 'species', y = 'petal_len', data = df)
```

```python
x = np.random.rand(100)
y = 10 * x + np.random.randn(100)
```

```python
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lrm = LinearRegression()
```

```python
X = x.reshape(-1, 1)
X.shape
```

```python
lrm.fit(X,y)
```

```python
x_fit = (np.linspace(-0.1, 1).reshape(-1, 1))
y_fit = lrm.predict(x_fit)
```

```python
plt.scatter(x,y)
plt.plot(x_fit, y_fit)
```
